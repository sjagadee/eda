## EDA With red wine data

The Red Wine Quality dataset is a widely used dataset in machine learning, particularly for classification and regression tasks. It consists of various chemical properties of red wine samples, along with a quality rating assigned by expert wine tasters. The dataset is sourced from the UCI Machine Learning Repository and is part of a study on the quality of red and white variants of the Portuguese “Vinho Verde” wine.

The attributes are (based on physiological test)
1) Alcohol
2) Malic acid
3) Ash
4) Alcalinity of ash  
5) Magnesium
6) Total phenols
7) Flavanoids
8) Nonflavanoid phenols
9) Proanthocyanins
10) Color intensity
11) Hue
12) OD280/OD315 of diluted wines
13) Proline
